```python
#upload video for pothole detection
import cv2
from ultralytics import YOLO

#loading trained model
model = YOLO("/kaggle/working/best.pt")

input_video_path = "/kaggle/input/pothole-video/input_video.mp4"
output_video_path = "/kaggle/working/output_with_detections.mp4"

#processing input video
cap = cv2.VideoCapture(input_video_path)
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps    = int(cap.get(cv2.CAP_PROP_FPS))

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

#frame-by-frame detection
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model.predict(source=frame, conf=0.5, imgsz=640, stream=True)

    for r in results:
        frame = r.plot()  

    out.write(frame)

cap.release()
out.release()
cv2.destroyAllWindows()

print(f"Output saved to: {output_video_path}!")
```

Real-Time Webcam Pothole Detection (for local machines)

This script enables real-time pothole detection using your laptop's webcam.
It draws bounding boxes and shows danger alerts if a pothole is detected too close.

> To run this, kindly download the notebook and `best.pt`, then run it locally with Python and OpenCV.


```python
#real-time pothole detection using webcam
import cv2
from ultralytics import YOLO

model = YOLO("best.pt")
cap = cv2.VideoCapture(0)

frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
max_area = frame_width * frame_height

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model.predict(source=frame, conf=0.5, imgsz=640, stream=True)

    for r in results:
        boxes = r.boxes.xyxy.cpu().numpy()

        for box in boxes:
            x1, y1, x2, y2 = map(int, box)
            area = (x2 - x1) * (y2 - y1)
            norm_area = min(area / max_area, 1.0)

            red = int(255 * norm_area)
            blue = int(255 * (1 - norm_area))
            color = (blue, 0, red)

            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            cv2.putText(frame, "pothole", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

            if norm_area > 0.15:
                cv2.putText(frame, "DANGER! Pothole Ahead!", (50, 50),
                            cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 0, 255), 3)

    cv2.imshow("Real-Time Pothole Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
